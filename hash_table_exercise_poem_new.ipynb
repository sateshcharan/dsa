```python
import csv

word_count = {}

with open ('./res/poem.txt', 'r') as file:
    for line in file:
        tokens = line.split(" ")
        for token in tokens:
            token = token.replace('\n', '')
            if token in word_count:
                word_count[token] += 1
            word_count[token] = 1
```
